```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
onebatch = pd.read_csv("../data/OneBatch3_FINAL2.csv")
```

```python
onebatch = onebatch.sort_values(by=["Donor", "Region"])
onebatch
```

```python
hc_lipids = onebatch[onebatch.SZ == "HC"].iloc[:, 9:].T.values
sz_lipids = onebatch[onebatch.SZ == "SZ"].iloc[:, 9:].T.values
```

```python
hc_pca.shape
```

```python
pca = PCA(n_components=2)

fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1)

hc_pca = pca.fit_transform(hc_lipids)
ax1.scatter(hc_pca[:, 0], hc_pca[:, 1])
ax1.set_xlabel(f"PC1, {pca.explained_variance_ratio_[0] * 100 :.1f} %")
ax1.set_ylabel(f"PC2, {pca.explained_variance_ratio_[1] * 100 :.1f} %")
ax1.set_title("PCA for healthy brains\n445 points\neach point is 1 lipid")

pca = PCA(n_components=2)
sz_pca = pca.fit_transform(sz_lipids)
ax2.scatter(sz_pca[:, 0], sz_pca[:, 1])
ax2.set_xlabel(f"PC1, {pca.explained_variance_ratio_[0] * 100 :.1f} %")
ax2.set_ylabel(f"PC2, {pca.explained_variance_ratio_[1] * 100 :.1f} %")
ax2.set_title("PCA for SZ brains\n445 points\neach point is 1 lipid")

plt.tight_layout()
```
